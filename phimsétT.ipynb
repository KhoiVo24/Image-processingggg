```python
import matplotlib.pyplot as plt
img=plt.imread('input/đenthui.jpg')
plt.figure(figsize=[15,5])
plt.imshow(img)
plt.title('ảnh mặt l',fontname='Tahoma',fontsize=20,color='Black')
# plt.grid() #đường caro trên hình
plt.axis() #hiển thị trục tung, hoành
plt.colorbar()
# plt.show('off')
```

```python
import numpy as np
import matplotlib.pyplot  as plt
plt.figure(figsize=(20,10))
r=img[:,:,0]
g=img[:,:,1]
b=img[:,:,2]
gay_lum=0.3*r+0.59*g+0.11*b
plt.subplot(121)
plt.imshow(img)
plt.title('ảnh gốc',fontname='Tahoma')
plt.subplot(122)
plt.imshow(gay_lum)
```
